### random sampling for test

```python
testSalvini = postDB_F_pol.loc[postDB_F_pol['Politician_x']=='MATTEO_SALVINI']['replyToText'].sample(n=10)
testSantanch = postDB_F_pol.loc[postDB_F_pol['Politician_x']=='DANIELA_SANTANCH']['replyToText'].sample(n=10)
testMeloni = postDB_F_pol.loc[postDB_F_pol['Politician_x']=='GIORGIA_MELONI']['replyToText'].sample(n=10)
testBoldrini = postDB_F_pol.loc[postDB_F_pol['Politician_x']=='LAURA_BOLDRINI']['replyToText'].sample(n=10)
testPillon = postDB_F_pol.loc[postDB_F_pol['Politician_x']=='SIMONE_PILLON']['replyToText'].sample(n=10)
testBellanova = postDB_F_pol.loc[postDB_F_pol['Politician_x']=='TERESA_BELLANOVA']['replyToText'].sample(n=10)

frames = [testSalvini, testSantanch,testMeloni,testBoldrini,testPillon,testBellanova]

#concatenate dataframes
test = pd.concat(frames, sort=False)
test.to_csv('test_valutazioni.csv')
test
```

```python
# creo una nuova variabile per ogni post con il 
z = ComPostDB_F_pol20['replyToText'].value_counts()
z1 = z.to_dict() #converts to dictionary
ComPostDB_F_pol20['Count_probl'] = ComPostDB_F_pol20['replyToText'].map(z1) 
ComPostDB_F_pol20 = ComPostDB_F_pol20.sort_values('Count_probl', ascending=False)
ComPostDB_F_pol20
```

```python

ComPostDB_F_pol20_166 = ComPostDB_F_pol20.loc[ComPostDB_F_pol20['Count_probl']>=2]
#ComPostDB_F_pol20_166

ComPostDB_F_pol20_166post = ComPostDB_F_pol20_166.drop_duplicates('replyToText')
#ComPostDB_F_pol20_166post
```

```python
# conto quanti post per ogni politico per fare un db di controllo bilanciato
a = ComPostDB_F_pol20_166post.groupby('Politician_x').count()
#a
```

```python
# creo la controparte bilanciata del db dei post con commenti problematici

ComPostDB_F_polNoHate = ComPostDB_F_pol.loc[ComPostDB_F_pol['Rating']<=19].drop_duplicates('replyToText')

b = ComPostDB_F_polNoHate.loc[ComPostDB_F_polNoHate['Politician_x'] == 'DANIELA_SANTANCH'].sample (31)
c = ComPostDB_F_polNoHate.loc[ComPostDB_F_polNoHate['Politician_x'] == 'GIORGIA_MELONI'].sample (37)
d = ComPostDB_F_polNoHate.loc[ComPostDB_F_polNoHate['Politician_x'] == 'LAURA_BOLDRINI'].sample (4)
e = ComPostDB_F_polNoHate.loc[ComPostDB_F_polNoHate['Politician_x'] == 'MATTEO_SALVINI'].sample (64)
f = ComPostDB_F_polNoHate.loc[ComPostDB_F_polNoHate['Politician_x'] == 'SIMONE_PILLON'].sample (24)
g = ComPostDB_F_polNoHate.loc[ComPostDB_F_polNoHate['Politician_x'] == 'TERESA_BELLANOVA'].sample (6)


frames = [b,c,d,e,f,g]
ComPostDB_F_polNoHate166 = pd.concat(frames, sort=False)
ComPostDB_F_polNoHate166
```
